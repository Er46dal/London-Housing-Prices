```python
df['year']=df['Month'].apply(lambda t:t.year)
```

```python
df.tail()
```

**3. Modeling**

We will define a function that will calculate a ratio of house prices, comparing the price of a house in 2018 to the price in 1998.

```python
ave_price=df.groupby(['London_Borough','year']).mean()
```

```python
ave_price.head()
```

```python
mean_price=ave_price.reset_index()
```

```python
mean_price.shape
```

```python
def create_price_ratio(d):
    price1998=float(d['Average_price'][d['year']==1998])
    price2018=float(d['Average_price'][d['year']==2018])
    ratio=[price2018/price1998]
    return ratio
```

```python
create_price_ratio(mean_price[mean_price['London_Borough']=='Barking & Dagenham'])
```

```python
ratios={}
```

```python
for b in mean_price['London_Borough'].unique():
    borough=mean_price[mean_price['London_Borough']==b]
    ratios[b]=create_price_ratio(borough)
print(ratios)    
```

```python
df_ratios=pd.DataFrame(ratios)
```

```python
df_ratios.head()
```

```python
df_ratios_T=df_ratios.T
```

```python
df_ratios_T.head()
```

```python
df_ratios_T.reset_index()
```

```python
df_ratios_T.rename(columns={'index':'Borough', 0:'2018'}, inplace=True)
top15=df_ratios_T.head(15)
ax=top15.sort_values(by='2018', ascending=False)
ax.plot(kind='bar')
plt.ylabel('Ratio')
plt.xlabel('Borough')
plt.title('London Housing Price Change (1998-2018)')
plt.show()
```
